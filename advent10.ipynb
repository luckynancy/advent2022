Read input

```python
input = []
with open('advent10.txt', encoding='utf8') as f:
    for line in f:
        x = line.split()
        input.append(x)
```

Part I

```python
x = 1
cycle = 0
ss = 0
for item in input:
    if cycle+1 in range(20,300,40):
            ss += x * (cycle + 1)
    
    cycle += 1
    if len(item)!=1:
        if cycle+1 in range(20,300,40):
            ss += x * (cycle + 1) 
        cycle += 1
        x += int(item[1])
        
print('Signal strength, {}'.format(ss))  
```

Part II

```python
X = [] # list will be used to collection x's

x = 1 # initial value 
for item in input: 
    X.append(x)
    if len(item)!=1: 
        X.append(x)     
        x += int(item[1])
        
crt = []
for i in range(0,240,40):
    crt.append(X[i:i+40])
```

```python
import pandas as pd
df = pd.DataFrame(crt) # CRT screen

for col in df.columns:
    for idx in df.index:
        val = df.loc[idx,col]
        if val in [col-1, col, col+1]:
            df.loc[idx,col] = "#"
        else:
            df.loc[idx,col] = '.'
```
